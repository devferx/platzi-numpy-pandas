# Filtrado con loc y iloc
Cuando queremos navegar por un dataFrame estas funciones permiten filtrar datos de manera más específica


```python
import pandas as pd
```

## .loc
Filtra según un label

```python
df_books = pd.read_csv('https://raw.githubusercontent.com/devferx/platzi-numpy-pandas/main/11_leer_archivos_csv_json_pandas/bestsellers-with-categories.csv')
df_books
```

Mostrar un rango de filas tomando en cuenta el start y el end

```python
df_books[0:4]
```

Filtrando por una columna

```python
df_books['Name']
```

Filtrando por varias columnas

```python
df_books[['Name', 'Author', 'Year']]
```

Filtrando por filas y columnas

```python
df_books.loc[0:4, ['Name', 'Author']]
```

Podemos modificar los valores de una columna específica del dataFrame

```python
df_books.loc[:, ['Reviews']] * -1
```

Filtrar datos que cumplan una condición determinada

```python
df_books.loc[:, ['Author']] == 'JJ Smith'
```

## .iloc
Filtra mediante indices


```python
df_books.iloc[:]
#muestra todos los datos del dataframe
```

Filtrar datos según los índices de las filas y las columnas

```python
df_books.iloc[:, 0:3]
# df_books.iloc[n:m, k:l]
# Muestra los datos de las columnas de n a m
# Muesta los datos de las filas con sus INDICES de k a l
```

Buscar un dato específico.

```python
df_books.iloc[1,3] * -1
#muestra el dato alojado en la fila 1 columna 3
```

```python
df_books.iloc[:2,2:]
```
